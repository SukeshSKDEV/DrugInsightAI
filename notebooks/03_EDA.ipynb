## EDA Summary

- The dataset was cleaned by removing missing values and exact duplicate records.
- Multiple bioactivity measurements for the same molecule were aggregated using the median pIC50.
- The final dataset contained 956 unique molecules.
- The pIC50 distribution was slightly right-skewed with a few biologically meaningful outliers that were retained.
- Molecular descriptors (LogP, Molecular Weight, HBA, and HBD) showed realistic distributions with some valid outliers.
- Correlation analysis revealed weak linear relationships between the molecular descriptors and pIC50, indicating that no single descriptor strongly predicts bioactivity.
- Scatter plot analysis showed clustered molecules and no clear linear relationship, suggesting complex nonlinear interactions.
- Based on the EDA findings, Random Forest Regressor was selected because it can model nonlinear relationships and feature interactions.
