```python
import requests
import pandas as pd
```

```python
target_name=input("Enter the Target Name : ")
```

```python
url=(
    "https://www.ebi.ac.uk/chembl/api/data/target/search.json"
    f"?q={target_name}"
)
```

```python
response=requests.get(url)
print(response.status_code)
```

```python
data=response.json()
```

```python
type(data)
```

```python
data.keys()
```

```python
selected_target=None

for target in data['targets']:
    
    if (
        target['organism']=="Homo sapiens"
        and target['target_type']=="SINGLE PROTEIN"
    ):
        selected_target=target
        break
```

```python
target=selected_target
```

```python
print(f"target name : {target['pref_name']}")
print(f"target ID : {target['target_chembl_id']}")
print(f"target organism : {target['organism']}")
```

```python
print(data['page_meta'])
```

```python
all_records=[]
```

```python
target_chembl_id=target['target_chembl_id']
```

```python
url=(
    "https://www.ebi.ac.uk/chembl/api/data/activity.json"
    f"?target_chembl_id={target_chembl_id}"
    "&standard_type=IC50"
    "&limit=100"
)
```

```python
response=requests.get(url)
data=response.json()
data['page_meta']
```

```python
page=1

while url:
    print(f"Downloading Page {page}")
    
    response=requests.get(url)
    data=response.json()
    all_records.extend(data['activities'])
    print(f"Records collected: {len(all_records)}")
    
    next_page=data["page_meta"]["next"]
    
    if next_page:
        url="https://www.ebi.ac.uk"+next_page
        page+=1
    else:
        url=None
print("Download Completed!")
```

```python
df=pd.DataFrame(all_records)
df.head()
```

```python
df.to_csv('../data/raw/activity.csv, index=False')
```
