## Building the DrugInsightAI Dataset

```python
url = ("https://www.ebi.ac.uk/chembl/api/data/activity.json"
      "?target_chembl_id=CHEMBL203"
      "&standard_type=IC50"
      "&limit=20")

response=requests.get(url)
print(response.status_code)
```

```python
data=response.json()
```

```python
type(data)
```

```python
data.keys()
```

```python
activities=data['activities']
activities[0]
```

```python
df=pd.DataFrame(activities)
df.head()
```

```python
df.columns
```

```python
df=df[['molecule_chembl_id','target_chembl_id','standard_type','standard_value','standard_units']]
df.head()
```

```python
df.to_csv('../data/raw/egfr_activity.csv, index=False')
```
